```python
import base64
import os
from google import genai
from google.genai import types


def generate():
    client = genai.Client(
        api_key=os.environ.get("GEMINI_API_KEY"),
    )

    model = "gemini-2.5-flash-preview-04-17"

    # --- 수정된 부분 시작 ---
    # 키보드로부터 사용자 입력을 받습니다.
    user_prompt = input("프롬프트 입력: ")
    # --- 수정된 부분 끝 ---

    contents = [
        types.Content(
            role="user",
            parts=[
                # --- 수정된 부분 시작 ---
                # 받은 사용자 입력을 text로 사용합니다.
                types.Part.from_text(text=user_prompt),
                # --- 수정된 부분 끝 ---
            ],
        ),
    ]
    generate_content_config = types.GenerateContentConfig(
        response_mime_type="text/plain",
    )

    print("\n응답:") # 응답 시작을 알리는 메시지 추가
    for chunk in client.models.generate_content_stream(
        model=model,
        contents=contents,
        config=generate_content_config,
    ):
        print(chunk.text, end="")
    print() # 스트리밍 완료 후 줄바꿈

if __name__ == "__main__":
    generate()
```

```python
import base64
import os
from google import genai
from google.genai import types

def generate_multi_turn_chat():
    """
    Gemini 모델과 멀티턴 대화를 수행하는 함수
    """
    client = genai.Client(
        api_key=os.environ.get("GEMINI_API_KEY"),
    )

    # 사용할 모델 지정
    model = "gemini-2.5-flash-preview-04-17"

    # 대화 기록을 저장할 리스트 초기화
    # 각 요소는 types.Content 객체이며, role("user" 또는 "model")과 parts(텍스트 등)를 가짐
    conversation_history = []

    # 모델 설정 (텍스트 응답만 받도록 설정)
    generate_content_config = types.GenerateContentConfig(
        response_mime_type="text/plain",
    )

    print("Gemini 챗봇 시작 (종료하려면 '안녕', 'exit', 'quit' 입력)")

    # 대화 루프 시작
    while True:
        # 키보드로부터 사용자 입력을 받습니다.
        user_prompt = input("나: ")

        # 종료 조건 확인
        if user_prompt.lower() in ["안녕", "exit", "quit"]:
            print("챗봇 종료.")
            break

        # 사용자의 입력을 대화 기록에 추가 (role="user")
        user_content = types.Content(
            role="user",
            parts=[types.Part.from_text(text=user_prompt)],
        )
        conversation_history.append(user_content)

        # Gemini 모델에게 대화 기록 전체를 전달하여 응답 요청
        print("봇: ", end="") # 봇의 응답 시작을 알림

        model_response_text = "" # 모델의 전체 응답을 누적할 변수

        try:
            # generate_content_stream 호출 시 전체 대화 기록(conversation_history) 전달
            for chunk in client.models.generate_content_stream(
                model=model,
                contents=conversation_history, # 전체 대화 기록 전달
                config=generate_content_config,
            ):
                # 스트리밍 응답 출력
                print(chunk.text, end="")
                # 전체 응답 텍스트 누적
                model_response_text += chunk.text

            print() # 스트리밍 완료 후 줄바꿈

            # 모델의 응답을 대화 기록에 추가 (role="model")
            # 응답이 비어있지 않은 경우에만 추가
            if model_response_text:
                 model_content = types.Content(
                    role="model",
                    parts=[types.Part.from_text(text=model_response_text)],
                )
                 conversation_history.append(model_content)
            else:
                print("(응답이 없습니다.)") # 모델이 응답을 보내지 않았을 경우

        except Exception as e:
            print(f"\nAPI 호출 중 오류 발생: {e}")
            # 오류 발생 시, 마지막에 추가했던 사용자 입력 제거 (선택사항)
            #conversation_history.pop()
            # break # 심각한 오류 시 종료

# 스크립트 실행 시 generate_multi_turn_chat 함수 호출
if __name__ == "__main__":
    generate_multi_turn_chat()
```
